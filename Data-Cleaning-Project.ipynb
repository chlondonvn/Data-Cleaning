# Handling Missing Info.:

```python
#The function below returns 'True' where there ARE missing values:
df.isna()
```

```python
#The sum of the missing values returns 
#0: when there are NO missing values and NUMBERS according to the sum of MISSING values

df.isna().sum()
```

```python
#To know with what type of data are working:

type(df.isna().sum())
```

```python
#Assigning missing values to a variable:

missing_values = df.isna().sum()
```

```python
#Filtering with '.loc.' in order to obtain only the columns with missing values:

missing_values.loc[missing_values.gt(0)]
```

```python
#To know what type of percentage of values are we misisng:

#This step is useful in order to clarify which columns we should drop.

missing_values[missing_values.gt(0)]/len(df)
```

```python
#Assigning columns with missing values to a variable in order to manipulate:

missing_values_percentage= missing_values[missing_values.gt(0)]/len(df)
```

```python
#Getting columns where missing values percentage is greater than 75%:

missing_values_percentage.loc[missing_values_percentage.gt(0.75)].index
```

```python
#Copying df and using df.drop() in order to eliminate the columns with high percentage of missing values:

df_copy1 = df.drop(columns=missing_values_percentage.loc[missing_values_percentage.gt(0.5)].index).copy()
```

```python
#Copy of data frame without columns: 

df_copy1
```

```python
#Defining a function in order to return columns with missing values to avoid repeating:

def columns_missing_value(df):
    return (df.isna().sum())[df.isna().sum() > 0]
```

```python
#Remaining columns with missing values:

columns_missing_value(df_copy1)

```
